```python
# Exam Answer Length Predictor System

print("======================================")
print("     EXAM ANSWER LENGTH PREDICTOR      ")
print("======================================\n")
```

```python
marks = int(input("Enter question marks (2 / 5 / 10): "))
qtype = input("Enter question type (theory / numerical): ").lower()
```

```python
if marks == 2 and qtype == "theory":
    points = "2 key points"
    words = 50
    diagram = "Not required"

elif marks == 5 and qtype == "theory":
    points = "5 clear points"
    words = 200
    diagram = "1 diagram recommended"

elif marks == 10 and qtype == "theory":
    points = "8 to 10 detailed points"
    words = 400
    diagram = "2 diagrams recommended"

elif marks == 2 and qtype == "numerical":
    points = "Step-by-step solution"
    words = 40
    diagram = "Not required"

elif marks == 5 and qtype == "numerical":
    points = "Formula + clear steps"
    words = 150
    diagram = "Not required"

elif marks == 10 and qtype == "numerical":
    points = "Detailed steps with explanation"
    words = 300
    diagram = "Not required"

else:
    print("Invalid input! Please enter correct values.")
    points = words = diagram = None
```

```python
if points is not None:
    print("\n-------- PREDICTED ANSWER FORMAT --------")
    print("Points to write       :", points)
    print("Approximate word count:", words)
    print("Diagram requirement   :", diagram)
```

```python
if points is not None:
    print("\nWriting Tips:")
    print("- Start with definition")
    print("- Write answers in points")
    print("- Underline important keywords")
    print("- Draw neat diagrams if required")
    print("- Manage time according to marks")
```
